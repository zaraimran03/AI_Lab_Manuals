```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, precision_score, recall_score
```

```python
df = pd.read_csv('/content/student_performance.csv')
print(df.head())

X = df[['Hours_Study', 'Hours_Sleep', 'Attendance']]
y = df['Marks']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# Plot
plt.scatter(y, y_pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Marks")
plt.show()

print("R² Score:", r2_score(y, y_pred))
print("Mean Squared Error:", mean_squared_error(y, y_pred))
```

```python
df = pd.read_csv('/content/pass_fail.csv')
print(df.head())

X = df[['Hours_Study', 'Hours_Sleep']]
y = df['Pass']

model = LogisticRegression()
model.fit(X, y)

# Predict probability
prob = model.predict_proba([[30, 6]])[0][1]
print("Probability of passing (30 hrs study, 6 hrs sleep):", prob)

# Plot decision regions
plt.scatter(df['Hours_Study'], df['Hours_Sleep'], c=y, cmap='bwr')
plt.xlabel("Hours Study")
plt.ylabel("Hours Sleep")
plt.title("Decision Boundary (Pass vs Fail)")
plt.show()
```

```python
df = pd.read_csv('/content/diabetes.csv')
print(df.head())

X = df[['BMI', 'Age', 'Glucose']]
y = df['Diabetic']

model = LogisticRegression()
model.fit(X, y)

y_pred = model.predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred))
print("Recall:", recall_score(y, y_pred))

pred = model.predict([[28, 45, 150]])
print("Prediction for patient:", "Diabetic" if pred[0]==1 else "Not Diabetic")
```

```python
df = pd.read_csv('/content/exam_comparison.csv')
print(df.head())

# Linear Regression for continuous score
X = df[['Hours_Study']]
y_score = df['Exam_Score']

lin_model = LinearRegression()
lin_model.fit(X, y_score)

print("Linear Regression predicted score (6 hrs):", lin_model.predict([[6]])[0])

# Logistic Regression for pass/fail
y_pass = df['Pass']
log_model = LogisticRegression()
log_model.fit(X, y_pass)

print("Logistic Regression predicted pass probability (6 hrs):", log_model.predict_proba([[6]])[0][1])

print("""
Explanation:
Linear Regression predicts continuous scores like exam marks.
Logistic Regression predicts probability for pass/fail (0 or 1).
Linear regression is unsuitable for classification because its output can go beyond 0–1 range.
""")
```
