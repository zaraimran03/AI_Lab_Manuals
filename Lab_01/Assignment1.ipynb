```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Student_ID', y='Marks_Math', data=data)
plt.title("Student_ID vs Marks_Math")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(data['Age'], bins=[16,17,18,19,20,21], edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='Marks_Math', y='Marks_Science', data=data, s=80)
sns.regplot(x='Marks_Math', y='Marks_Science', data=data, scatter=False, ci=None)
plt.title("Math vs Science Marks")
plt.show()
```
